```python
import numpy as np
import pandas as pd
from collections import Counter
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("C:\\Projects\\Resume_checker\\data\\UpdatedResumeDataSet.csv")
```

```python
dataset.head(5)
```

```python
def Keywords(resume,pattern):
    matches=[]
    for text in resume:
        matches.extend(re.findall(pattern,str(text),flags=re.IGNORECASE))
    return matches
```

```python
skills_pattern=r"(Python|R|SQL|Tableau|SAP HANA|pandas|numpy|scikit-learn|TensorFlow|PyTorch)"
skills=Keywords(dataset["Resume"],skills_pattern)
skills_count=Counter(skills)

plt.figure(figsize=(12,6))
plt.bar(skills_count.keys(),skills_count.values())
plt.title("Types of Skills Across All Resumes")
plt.xlabel("Skills")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
interests_pattern=r"(Deep Learning|Control Systems|AI|Machine Learning|Data Analysis|Big Data)"
interests=Keywords(dataset["Resume"],interests_pattern)
interests_count=Counter(interests)
plt.figure(figsize=(12,6))
plt.bar(interests_count.keys(),interests_count.values())
plt.title("Area of Interests")
plt.xlabel("Interests")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
education_pattern=r"(B\.E\.|M\.S\.|B\.Tech|M\.Tech|Ph\.D|MBA)"
education=Keywords(dataset["Resume"],education_pattern)
education_count=Counter(education)
plt.figure(figsize=(12,6))
plt.bar(education_count.keys(),education_count.values())
plt.title("Different Education Backgrounds")
plt.xlabel("Education")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=dataset, x='Category', palette='viridis')

plt.title("Distribution of Categories", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=60)  
plt.tight_layout()
```
